**3.24: Slicing Rows lazily**

**3.25: Slicing Lexicographically:**

Slicing lexicographically The .loc indexer typically selects data based on the exact string label of the index. However, it also allows you to select data based on the lexicographic order of the values in the index. Specifically, .loc allows you to select all rows with an index lexicographically using slice notation.
<br>This works only if the index is sorted.

````
try:
    print(canada_copy['Bu' : 'Ca'])
except Exception as e:
    print(type(e), e)
````

````
canada_copy.sort_index(ascending=True, inplace=True)
canada_copy.head()
````

Now let's run the command again

````
try:
    print(canada_copy['Bu' : 'Ca'])
except Exception as e:
    print(e)
````

### **Part 3.5: Speeding up Scalar selection**

Both the `.iloc` and `.loc` indexers are capable of selecting a single element, a scalar value, from a Series or DataFrame. However, there exist the indexers,`.iat` and `.at`, which respectively achieve the same thing at faster speeds.

Like `.iloc`, the `.iat` indexer uses integer location to make its selection and must be passed two integers separated by a comma. Similar to `.loc`, the `.at` index uses labels to make its selection and must be passed an index and column label separated by a comma.

This recipe is valuable if computational time is of utmost importance. It shows the performance improvement of `.iat` and `.at` over `.iloc` and `.loc` when using Scalar selection.

**3.51: Selecting the migration figure for Nigeria in 2010 using `.loc` and `.at`**

````
canada_copy.loc['Nigeria','2010']
````

Doing the same with `.at`

````
canada_copy.at['Nigeria','2010']
````

**3.52: Use the `%timeit` magic command to find the difference in speed:**

````
%timeit canada_copy.loc['Nigeria','2010']
````

````
%timeit canada_copy.at['Nigeria','2010']
````

So we get 4.26 microseconds for `.at` and 7.57 microseconds for `.loc`.

### **Part 3.6: Boolean Selections**

**3.61: Filtering based on a criteria**

To filter the dataframe based on a condition, we simply pass the condition as a boolean vector. For example, Let's filter the dataframe to show the data on African countries (Continent = Asia).

**Step 1:** We could either sort it and return a boolean series of True and False values like this:
<br>Returning True for African Countries and False otherwise.

````
canada_copy['Continent'] == 'Africa'
````

**Step 2:** Or simply pass the boolean sorting condition in our data frame to select the countries that are true

````
canada_copy[canada_copy.Continent == 'Africa'].head()  # Returning the first 5 African Countries
````

**3.62: Multiple Booleans with pandas square brackets using ampersand(&).**

The ampersand sign (&) denotes 'and' like selecting two conditions together.<br> Let's select countries only in Africa that have atleast 50000 total migrations to Canada

````
canada_copy[(canada_copy['Total'] >= 50000) & (canada_copy['Continent'] == 'Africa')]
````

So as we specified in our selection, we can see The Countries in Africa that have atleast 50000 total migrations to Cananda.

**3.63: Multiple Booleans with pandas square brackets using the Vertical Bar(|).**

The Vertical Bar sign (|) denotes 'or', like selecting either of two conditions or both together.<br> Let's select the countries that are either in Northern Africa or Northern America that migrated to Canada during the period.

````
canada_copy[(canada_copy['Region'] == 'Northern Africa') | (canada_copy['Region'] == 'Northern America')]
````

So we can see the countries that are either from Northern Africa or Northern America above.


**3.64 Multiple Booleans on DataFrame with np.logical_and condition**

We can also use the `np.logical` condition from the numpy library to subset a pandas DataFrame. Let's use this to select Asian countries that had no more than 50 migrations to Canada in the year 2010

````
canada_copy[np.logical_and(canada_copy['Continent'] == 'Asia', canada_copy['2010'] <= 50)]
````

**3.65 Multiple Booleans on DataFrame with np.logical_or condition**

We can also use the `np.logical_or` condition from the numpy library to subset a pandas DataFrame. Let's use this to select countries in either Northern Asia or Southern Africa

````
canada_copy[np.logical_or(canada_copy['Region']=='Southern Asia', canada_copy['Region']=='Southern Africa')]
````

### **Part 3.6: Binning**

This is a method for Data pre-processing. Binning involves grouping data values into bins.
For example we can bin total migrations into low, medium, high and very high bins. Sometimes binning can improve accuracy of predictive models As well as make analysing data easier.

**Step One: Defining the bin width**

Since we want 4 bins from the Total column, our binwidth should be 4 in order to get a proper all equal width

````
# Let's set the binwidth from the range of values in the Total Column

bin_width = max(canada_copy.Total) //4
print(bin_width)
````

**Step Two: Creating The Bin Array**

From min value to max value using a range and also passing binwidth as the step



````
bin = list(range(0, max(canada_copy.Total)+1, bin_width))
bin
````

**Step Three: Creating labels for each Bin_**

We now create a simple list of labels:- Low, Medium, High, V_High

````
labels = ['Low', 'Medium', 'High', 'V_High']
````

**Step Four:Finally we create a new column with the defined bins as its values using pd.cut() function**

````
canada_copy['Migration_Rate'] = pd.cut(canada_copy.Total, bin, labels=labels, include_lowest=True)
canada_copy.head(10)
````
